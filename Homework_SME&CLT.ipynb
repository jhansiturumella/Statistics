```python
#x = [1,2,5,8,5,3,9]
#x_arr = np.array(x)
x_arr = np.random.rand(20)
print(x_arr)
mean = np.mean(x_arr)
sd = np.std(x_arr)
SME = sd/math.sqrt(len(x_arr))
print('Mean is {}'.format(SME))
```

## Sample Dsirtibution #2 as per CLT

```python
samples = [np.mean(random.choices(s, k=100)) for _ in range(500)] # Taking mean 100 times from 500 sample population as per Central Limit Theorem

plt.hist(samples)
plt.show()
```
